```python
import numpy as np
import matplotlib.pyplot as plt

dataset = np.array([
     [5.4, 2.5, 75],
    [5.2, 2, 72], #1 kg yem narxi, yoshi, narxi
    [5, 4, 68],
     [4.8, 3, 69],
    [4, 3, 67],


    ])
x= dataset[:,0]
y = dataset[:,-1]

print(x)
```

```python
plt.figure(figsize=(8,4))
plt.scatter(x,y,marker="s", color="r")
plt.xlabel("yem narxi")
plt.ylabel("go'sht narxi")
plt.title("nimadir")
plt.grid()
plt.show()
```

```python
p = np.polyfit(x,y,2)
reg_model = np.poly1d(p)
yNew = reg_model(x)
```

```python
plt.figure(figsize=(8,4))
plt.subplot(1,2,1)
plt.scatter(x,y,marker="s", color="r", label = "original data")
plt.subplot(1,2,2)
plt.plot(x,yNew, color = "b", label = "reg model")
plt.xlabel("yem narxi")
plt.ylabel("go'sht narxi")
plt.title("nimadir")
plt.grid()
plt.legend()
plt.show()
```
